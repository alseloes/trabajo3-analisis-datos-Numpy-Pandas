# ANÁLISIS DE DATOS: NUMPY & PANDAS

```python
# ============================================================
# 1. Importar librerías
# ============================================================

import pandas as pd
import numpy as np
```

```python
# ============================================================
# 2. Cargar los datos
# ============================================================

ventas = pd.read_csv("/workspace/sales.csv")
inventarios = pd.read_csv("/workspace/inventories.csv")
satisfaccion = pd.read_csv("/workspace/satisfaction.csv")
```

```python
# ============================================================
# 3. Limpieza de datos
# ============================================================

ventas = ventas.dropna()
inventarios = inventarios.dropna()
satisfaccion = satisfaccion.dropna()

print("===================================")
print("INFORMACIÓN DE LOS DATAFRAMES")
print("===================================")

print("\nVentas")
print(ventas.info())

print("\nInventarios")
print(inventarios.info())

print("\nSatisfacción")
print(satisfaccion.info())
```

```python
# ============================================================
# 4. Exploración de datos
# ============================================================

print("\n===================================")
print("ANÁLISIS DE VENTAS")
print("===================================")

# Calcular ingresos por registro

ventas["Total_Ventas"] = (
    ventas["Cantidad_Vendida"] *
    ventas["Precio_Unitario"]
)

print("\nDataFrame de ventas")
print(ventas.head())

# ------------------------------------------------------------
# Ventas totales por producto
# ------------------------------------------------------------

ventas_producto = (
    ventas
    .groupby("Producto")["Cantidad_Vendida"]
    .sum()
)

print("\nVentas totales por producto")
print(ventas_producto)

# ------------------------------------------------------------
# Ventas totales por tienda
# ------------------------------------------------------------

ventas_tienda = (
    ventas
    .groupby("ID_Tienda")["Cantidad_Vendida"]
    .sum()
)

print("\nVentas totales por tienda")
print(ventas_tienda)

# ------------------------------------------------------------
# Ingresos por tienda
# ------------------------------------------------------------

ingresos_tienda = (
    ventas
    .groupby("ID_Tienda")["Total_Ventas"]
    .sum()
)

print("\nIngresos por tienda")
print(ingresos_tienda)

# ------------------------------------------------------------
# Estadísticas descriptivas
# ------------------------------------------------------------

print("\nResumen estadístico")
print(ventas["Total_Ventas"].describe())

print("\nMediana (Pandas)")
print(ventas["Total_Ventas"].median())


# ------------------------------------------------------------
# Promedio por tienda
# ------------------------------------------------------------

promedio_tienda = (
    ventas
    .groupby("ID_Tienda")["Total_Ventas"]
    .mean()
)

print("\nPromedio de ventas por tienda")
print(promedio_tienda)


# ------------------------------------------------------------
# Si existe la columna Categoría
# ------------------------------------------------------------

if "Categoria" in ventas.columns:

    promedio_categoria = (
        ventas
        .groupby(["ID_Tienda", "Categoria"])["Total_Ventas"]
        .mean()
    )

    print("\nPromedio por tienda y categoría")
    print(promedio_categoria)

else:

    print("\nEl archivo no contiene la columna 'Categoria'.")
```

```python
# ============================================================
# 5. Análisis de inventarios
# ============================================================

print("\n===================================")
print("ANÁLISIS DE INVENTARIOS")
print("===================================")

inventario_ventas = pd.merge(
    inventarios,
    ventas,
    on=["ID_Tienda", "Producto"]
)

inventario_ventas["Rotacion"] = (
    inventario_ventas["Cantidad_Vendida"] /
    inventario_ventas["Stock_Disponible"]
)

inventario_ventas["Porcentaje_Vendido"] = (
    inventario_ventas["Cantidad_Vendida"] /
    inventario_ventas["Stock_Disponible"]
)

print("\nRotación de inventarios")
print(
    inventario_ventas[
        [
            "ID_Tienda",
            "Producto",
            "Stock_Disponible",
            "Cantidad_Vendida",
            "Rotacion"
        ]
    ]
)

# ------------------------------------------------------------
# Tiendas con inventario crítico
# ------------------------------------------------------------

criticos = inventario_ventas[
    inventario_ventas["Porcentaje_Vendido"] < 0.10
]

print("\nTiendas con inventario crítico")
print(criticos)

# ------------------------------------------------------------
# Rotación promedio por tienda
# ------------------------------------------------------------

rotacion_tienda = (
    inventario_ventas
    .groupby("ID_Tienda")["Rotacion"]
    .mean()
)

print("\nRotación promedio por tienda")
print(rotacion_tienda)
```

```python
# ============================================================
# 6. Satisfacción del cliente
# ============================================================

print("\n===================================")
print("SATISFACCIÓN DEL CLIENTE")
print("===================================")

analisis = pd.merge(
    satisfaccion,
    ingresos_tienda.reset_index(),
    on="ID_Tienda"
)

print("\nRelación satisfacción e ingresos")
print(analisis)

baja_satisfaccion = analisis[
    analisis["Satisfacción_Promedio"] < 60
]

print("\nTiendas con satisfacción menor al 60%")
print(baja_satisfaccion)

print("\nRECOMENDACIONES")

for tienda in baja_satisfaccion["ID_Tienda"]:
    print(f"""
Tienda {tienda}
- Mejorar la atención al cliente.
- Capacitar al personal.
- Revisar el inventario disponible.
- Implementar promociones.
- Analizar las causas de la baja satisfacción.
""")
```

```python
# ============================================================
# 7. Operaciones con NumPy
# ============================================================

print("\n===================================")
print("OPERACIONES CON NUMPY")
print("===================================")

ventas_numpy = ventas["Total_Ventas"].to_numpy()

mediana = np.median(ventas_numpy)
desviacion = np.std(ventas_numpy)

print("\nMediana")
print(mediana)

print("\nDesviación estándar")
print(desviacion)
```

```python
# ============================================================
# Simulación de ventas futuras
# ============================================================

np.random.seed(42)

proyeccion = np.random.normal(
    loc=np.mean(ventas_numpy),
    scale=np.std(ventas_numpy),
    size=12
)

print("\nProyección de ventas futuras")
print(proyeccion)

print("\nEstadísticas de la proyección")

print("Promedio :", np.mean(proyeccion))
print("Mediana  :", np.median(proyeccion))
print("Máximo   :", np.max(proyeccion))
print("Mínimo   :", np.min(proyeccion))
print("Desv.Est.:", np.std(proyeccion))
```
